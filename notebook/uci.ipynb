```python
from ucimlrepo import fetch_ucirepo, list_available_datasets
from sklearn import svm
import numpy as np
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
```

```python
list_available_datasets()
heart_disease = fetch_ucirepo(name='Wine Quality')
```

```python
X = heart_disease.data.features
y = heart_disease.data.targets
```

```python
X
```

```python
X_clean = imp.fit_transform(X)
```

```python
X_clean.shape
```

```python
np.mean(X_clean, axis=0)
```

```python
np.cov(X_clean, rowvar=False)
```

```python
regr = svm.SVR()
regr.fit(X_clean, y)
```

```python
import scipy
m, cov = np.mean(X_clean, axis=0), np.cov(X_clean, rowvar=False)
B = scipy.linalg.cholesky(cov)
```

```python
B
```

```python

B_inv = scipy.linalg.lapack.dtrtri(B)
```

```python
np.dot(B_inv[0], B)
```

```python
B_inv
```

```python
X_clean_norm = np.dot(X_clean - m, B_inv[0])
```

```python
np.cov(X_clean_norm, rowvar=False)
```
