```python
import geoopt
import torch
```

```python
euclidean = geoopt.manifolds.Euclidean()
spd = geoopt.manifolds.SymmetricPositiveDefinite()
manifold = geoopt.manifolds.ProductManifold((euclidean, 2), (spd, (2,2)))
```

```python
manifold = geoopt.manifolds.ProductManifold((euclidean, 2), (euclidean, (2,2)))
```

```python
t = torch.zeros(6)
t[2], t[5] = 1., 1.
```

```python
mani_tensor = geoopt.ManifoldTensor(t, manifold=manifold)
```

```python
param = geoopt.ManifoldParameter(mani_tensor)
```

```python
manifold.take_submanifold_value(param, 0), manifold.take_submanifold_value(param, 1)
```

```python
def loss(input):
    print(input)
    target = 4*torch.ones(6)
    target[3], target[4] = 0, 0
    return input.dist(target)
```

```python
optimizer = geoopt.optim.RiemannianSGD((param,), lr=1e-2)
```

```python
for iteration in range(100):
    optimizer.zero_grad()
    output = loss(param)
    output.backward()
    optimizer.step()
```
